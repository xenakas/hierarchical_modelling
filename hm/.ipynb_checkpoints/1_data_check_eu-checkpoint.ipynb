### 1. Данные для ЕС

```python
import pandas as pd
import numpy as np
```

[National accounts - main GDP aggregates and related indicators](https://ec.europa.eu/eurostat/statistics-explained/index.php/National_accounts_-_main_GDP_aggregates_and_related_indicators#Further_analysis_of_main_GDP_aggregates)

```python
eu28_codes = pd.read_html('http://publications.europa.eu/code/pdf/370000en.htm')[2].iloc[1:29,3].values
eu28_codes
```

Cуммарный ВВП для 28 стран (евростат)

```python
eu28_gdp = pd.read_csv('data/eu_gdp_eu28.csv', index_col=0, header=[0])
# eu28_gdp.iloc[20:,]
```

ВВП для каждой из 28 стран:

```python
eu_gdp_by_country = pd.read_csv('data/eu_gdp_by_country.csv', index_col=0, header=[0])
eu28_gdp_by_country = eu_gdp_by_country[eu28_codes].dropna(axis=0)
```

ВВП для каждой из 28 стран взятый из колонки  'TOTAL' 

```python
eu_a10_total = pd.read_csv('data/eu_a10_total.csv', index_col=0, header=[0])
eu28_a10_total = eu_a10_total[eu28_codes].dropna(axis=0)
# eu28_a10_total
```

ВВП для каждой из 28 стран для каждой из 12 отраслей

[List of NACE codes](http://ec.europa.eu/competition/mergers/cases/index/nace_all.html)

```python
eu_a10 = pd.read_csv('data/eu_a10.csv', index_col=0, header=[0])
```

```python
eu28_ind_codes = []
for i in list(eu28_codes):
    for j in list(['A', 'B', 'C', 'F', 'G', 'J', 'K', 'L', 'M', 'O', 'R']):
        eu28_ind_codes.append(i+j)
```

```python
eu28_ind_codes.remove('BEC')
```

```python
eu28_a10 = eu_a10[eu28_ind_codes].dropna(axis=0)
```

Разница между суммарным ВВП для 28 стран (евростат) и суммой ВВП для каждой из 28 стран (не больше 0.12% ).

```python
eu28_gdp.iloc[20:,].plot()
eu28_gdp_by_country.sum(axis=1).plot();
eu28_a10_total.sum(axis=1).plot();
```

```python
eu28_gdp.iloc[20:,].sub(eu28_gdp_by_country.sum(axis=1), axis=0).plot();
```

```python
eu28_gdp.iloc[20:,].sub(eu28_gdp_by_country.sum(axis=1), axis=0).apply(np.abs).div(eu28_gdp.iloc[20:,]).plot();
```

Разница между суммарным ВВП для 28 стран (евростат) и суммой ВВП для каждой из 28 стран взятой из колонки  'TOTAL' = Taxes less subsidies on products

```python
eu28_gdp.iloc[20:,].sub(eu28_a10_total.sum(axis=1), axis=0).plot();
```

```python
eu28_gdp.iloc[20:,].sub(eu28_a10_total.sum(axis=1), axis=0).div(eu28_gdp.iloc[20:,]).plot();
```

Разница между суммарным ВВП для 28 стран (евростат) и суммой ВВП для каждой из 28 стран по каждой из 12 отраслей

```python
eu28_gdp.iloc[20:,].sub(eu28_a10.sum(axis=1), axis=0).plot();
```

```python
eu28_gdp.iloc[20:,].sub(eu28_a10.sum(axis=1), axis=0).div(eu28_gdp.iloc[20:,]).plot();
```

```python
allcol = eu28_a10.columns.values
allcol_c = [i[:-1] for i in allcol]
allcol_i = [i[2] for i in allcol]
```

```python
allcol_c75 = [i for i in allcol_c for j in range(75)]
allcol_i75 = [i for i in allcol_i for j in range(75)]
len(allcol_c75)
```

```python
eu28_a10.plot(legend=False);
```

```python
unst = pd.DataFrame(eu28_a10.unstack())
unst['geo'] = allcol_c75
unst['ind'] = allcol_i75
unst['time'] = unst.index.get_level_values(1)
```

```python
eu_28_summed = unst.set_index(['geo','ind', 'time']).groupby(level=[0, 2]).sum().unstack(level=0)
```

```python
eu28_a10_total.head()
```

```python
eu_28_summed.head()
```

```python
eu28_C = unst[unst['ind']=='C'].set_index(['geo','time']).iloc[:,0].unstack(level=0)
```

Видимо, "Construction" не учитывается при подсчете ВВП

```python
eu28_a10_total.sub(eu_28_summed, axis = 0, level = 1).add(eu28_C).sum(axis=1).plot();
```

```python
eu28_a10_total.sub(eu_28_summed, axis = 0, level = 1).add(eu28_C).diff(4).sum(axis=1).plot();
```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(eu28_a10_total.sub(eu_28_summed, axis = 0, level = 1).add(eu28_C).diff(4).sum(axis=1));
```

Запишем файлы для `hts`, дающие в сумме `eu28_gdp`

1. по странам - оставим как есть
2. по a10 - уберем колонки с Construction, добавим столбик с 'Taxes less subsidies on products'

```python
eu28_new_codes = []
for i in list(eu28_codes):
    for j in list(['A', 'B', 'F', 'G', 'J', 'K', 'L', 'M', 'O', 'R']):
        eu28_new_codes.append(i+j)
```

```python
eu28_gdp.sub(eu_a10[eu28_new_codes].dropna(axis=0).sum(axis=1), axis=0).plot()
```

```python
eu28_gdp.sub(eu_a10[eu28_new_codes].dropna(axis=0).sum(axis=1), axis=0).sub(eu28_gdp.iloc[20:,].sub(eu28_a10_total.sum(axis=1), axis=0), axis = 0).plot()
```

```python
hts_a10 = eu_a10[eu28_new_codes].dropna(axis=0)
hts_a10['ZTX']=eu28_gdp.sub(eu_a10[eu28_new_codes].dropna(axis=0).sum(axis=1), axis=0).dropna(axis=0)
hts_a10.to_csv("hts_eu_a10.csv")
```
