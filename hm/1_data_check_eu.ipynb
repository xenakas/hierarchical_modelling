```python
import pandas as pd
import numpy as np
```

```python
eu28_codes = pd.read_html('http://publications.europa.eu/code/pdf/370000en.htm')[2].iloc[1:29,3].values
```

```python
eu28_gdp = pd.read_csv('data/eu_gdp_eu28.csv', index_col=0, header=[0])
# eu28_gdp.iloc[20:,]
```

```python
eu_gdp_by_country = pd.read_csv('data/eu_gdp_by_country.csv', index_col=0, header=[0])
eu28_gdp_by_country = eu_gdp_by_country[eu28_codes].dropna(axis=0)
```

```python
eu28_by_a10 = pd.read_csv('data/eu_a10_eu28.csv', index_col=0, header=[0])
# eu28_by_a10.iloc[20:,]
```

```python
eu_a10_total = pd.read_csv('data/eu_a10_total.csv', index_col=0, header=[0])
eu28_a10_total = eu_a10_total[eu28_codes].dropna(axis=0)
# eu28_a10_total
```

```python
eu_a10 = pd.read_csv('data/eu_a10.csv', index_col=0, header=[0])
```
