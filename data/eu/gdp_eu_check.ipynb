```python
df = pd.read_csv('/home/xenakas/Desktop/Git/hierarchical_modelling/gdp_eu.csv')

del df['Unnamed: 0']

pivot_gdp = pd.pivot_table(df, index='time', columns=['geo','nace_r2'], values = 'values')

pivot_gdp.to_csv("gdp_geo_and_nace.csv")

pivot_gdp 
```

```python
pivot_gdp_geo = pd.pivot_table(df, index='time', columns='geo', aggfunc=np.sum) 

pivot_gdp_geo.to_csv("gdp_geo.csv")

pivot_gdp_geo
```
