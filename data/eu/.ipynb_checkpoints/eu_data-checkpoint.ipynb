```python
import pandas as pd
import numpy as np
```

```python
df_gdp = pd.read_csv('a10_by_country_stacked.csv', index_col = 0, header = None ) #  header=[0,1], )
```

```python
df_gdp.iloc[0,:].unique()
```

```python
df_gdp.iloc[1,:].unique()
```

```python
df_gdp.iloc[1,:] = [x[0] for x in df_gdp.iloc[1,:]]
```

```python
df_gdp.iloc[1,:].unique()
```

```python
df_gdp.columns = np.array(df_gdp.iloc[0,:].values)   + np.array(df_gdp.iloc[1,:].values)
df_gdp = df_gdp.drop(['geo', 'nace_r2', 'time'], axis=0)
```

```python
df_gdp.describe().transpose().groupby('count').count()
```

```python
df_gdp.to_csv("renamed_a10.csv")
df_gdp.dropna(axis=0).to_csv("renamed_a10_short.csv")
```
