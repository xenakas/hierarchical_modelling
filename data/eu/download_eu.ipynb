```r
library(eurostat)
library(lubridate)
```

```r
a10 <- get_eurostat(id="namq_10_a10", time_format="date")
#a10 <- label_eurostat(a10)
head(a10)
```

```r
unique(a10$s_adj) 
a10 <- subset(a10, a10$s_adj=='NSA') # Unadjusted data (i.e. neither seasonally adjusted nor calendar adjusted data)
```

```r
unique(a10$na_item) 
a10 <- subset(a10, a10$na_item=='B1G') #Value added, gross
```

```r
unique(a10$unit, incomparables = FALSE)
a10 <- subset(a10, a10$unit=='CP_MEUR') #Current prices, million euro
```

```r
a10$unit <- NULL
a10$s_adj <- NULL
a10$na_item <- NULL
```

```r
head(a10)
```

```r
unique(a10$geo, incomparables = FALSE)
```

```r
write.csv(a10, file = "eu_gdp_a10_unstacked.csv")
```
