```python
import pandas as pd
import numpy as np
```

```python
df_gdp = pd.read_csv('naus_unlabelled.csv')
del df_gdp['Unnamed: 0']

df_gdp['series_id'] = [x[:-5] for x in df_gdp['series_id']]

ids = df_gdp['series_id'].unique()
len(ids)


ids

df_gdp['state'] = [df_gdp['series_id'][x][:2] for  x  in range(len(df_gdp['series_id']))]
df_gdp['ind'] = [df_gdp['series_id'][x][2:] for  x  in range(len(df_gdp['series_id']))]

df_gdp = df_gdp.replace('', 'TOTAL')

df_gdp = df_gdp[df_gdp['ind'] != 'TOTAL']

df_gdp = df_gdp[df_gdp['state'] != 'US']


del df_gdp['series_id']

df_gdp.date = pd.to_datetime(df_gdp.date)

#ids.reshape(60,24).tolist()
```

```python
pivot_gdp = pd.pivot_table(df_gdp, index='date', columns= ['state', 'ind'], values = 'value')
pivot_gdp

#pivot_gdp.to_csv("gdpus_unlabelled.csv")

```

```python
import matplotlib.pyplot as pp

pp.figure(figsize = (18, 8))
pp.plot(pivot_gdp)
```

```python
pivot_geosum = pd.pivot_table(df_gdp, index='date', columns='state', aggfunc=np.sum) 

pivot_geosum
pp.figure(figsize = (30, 20))
pp.plot(pivot_geosum)
pp.legend(df_gdp.state.unique())

time = pp.gca()
time.set_xticks(pd.date_range(start=min(df_gdp.date), end=max(df_gdp.date), freq = '3Y'))
time.set_xticklabels(np.arange(2005,2020,3), rotation=20)

```
