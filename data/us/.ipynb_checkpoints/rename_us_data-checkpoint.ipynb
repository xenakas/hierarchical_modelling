```python
import pandas as pd
import numpy as np
```

```python
df_gdp = pd.read_csv('naus_unlabelled.csv')
del df_gdp['Unnamed: 0']
```

```python
df_gdp['series_id'] = [x[:-5] for x in df_gdp['series_id']] # del rqgsp
```

```python
ids = df_gdp['series_id'].unique()
ids
```

```python
df_gdp['state'] = [df_gdp['series_id'][x][:2] for  x  in range(len(df_gdp['series_id']))]
df_gdp['ind'] = [df_gdp['series_id'][x][2:] for  x  in range(len(df_gdp['series_id']))]
```

```python
df_gdp
```

```python
df_gdp['state'].unique().size
```

```python
df_gdp['state'].unique()
```

```python
df_gdp = df_gdp.replace('', 'TOTAL')
```

```python
df_gdp = df_gdp[df_gdp['ind'] != 'TOTAL']
```

```python
del df_gdp['series_id']
df_gdp['ind'] = df_gdp['ind'].replace('PI', 'PIT')
```

```python
df_gdp['ind'] = [df_gdp['ind'][x][:3] for  x  in range(len(df_gdp['ind']))]
df_gdp['ind'].unique()
```

```python
df_gdp_total = df_gdp[df_gdp['ind'] == 'TOT']
del df_gdp_total['ind']
df_gdp_total
```

```python
df_gdp_total = pd.pivot_table(df_gdp_total, index='date', columns= 'state', values = 'value')
df_gdp_total.to_csv("new_total_ind.csv")
```

```python
df_gdp['index'] = df_gdp['state'] + df_gdp['ind']
del df_gdp['state']
del df_gdp['ind']
df_gdp
```

```python
pivot_gdp = pd.pivot_table(df_gdp, index='date', columns= 'index', values = 'value')
pivot_gdp.to_csv("gdp_us_state_ind.csv")
```
