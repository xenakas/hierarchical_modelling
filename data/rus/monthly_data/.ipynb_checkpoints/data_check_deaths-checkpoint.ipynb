#### 1

```python
change = pd.read_csv('well_change.csv', header = None, index_col=0, parse_dates=True)
```

```python
ind   = ['область', ' ', 'округ', 'автономный',  'автономная', 'край', 'республика', 'Республика', 
         'АО', 'г.', 'Город', 'федерального', 'значения', ' ', '.']
for stro in ind:
    change.iloc[0,:]  = [i.replace(stro, '') for i in change.iloc[0,:]]
dictionary = {"МосквастолицаРоссийскойФедерациигород":"мосг", "Красноярский":"края",  "Саха(Якутия)" :"якут", "Алтайский":"алтк" }
for key in dictionary.keys():
    change.iloc[0,:]  = [i.replace(key, dictionary[key]) for i in change.iloc[0,:]]
change.iloc[0,:] = change.iloc[0,:].apply(lambda x: x.upper()[:4])
```

```python
change.iloc[0,:].unique()
```

```python
dictionary_types = {"Рожденных":"B", "Умерших":"D"}

for key in dictionary_types.keys():
    change.iloc[1,:]  = [i.replace(key, dictionary_types[key]) for i in change.iloc[1,:]]

change.columns = change.iloc[0,:] + change.iloc[1,:]

change = change.drop(index = change.index[[0,1]])
```

```python
change.to_csv("hts_change.csv")
```
